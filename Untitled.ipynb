```python
import pandas as pd
import numpy as np
import matplotlib as plt
import seaborn as sns
```

```python
Exploratory Data Analysis
```

```python
Airport Name
```

```python
sales_itm = df.groupby(['Airport: Name'], as_index=False)['Wildlife: Number Struck Actual'].sum().sort_values(by='Wildlife: Number Struck Actual', ascending=False).head(10)
sns.set(rc={'figure.figsize':(30,5)})
ax = sns.barplot(x = 'Airport: Name',y= 'Wildlife: Number Struck Actual' ,data = sales_itm)
for bars in ax.containers:
    ax.bar_label(bars)
```
